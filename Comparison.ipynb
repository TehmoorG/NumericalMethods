# ACSE Functions analysis

## Exp Function

```python
from acsefunctions.taylor import exp, sin, cos, tan
import numpy as np
from scipy.special import gamma as scipy_gamma
from scipy.special import jv as scipy_bessel_j
import matplotlib.pyplot as plt
from time import time

x_val = 2
custom_exp_val = exp(x_val,N=20)
numpy_exp_val = np.exp(x_val)

print(f"Custom exp({x_val}) = {custom_exp_val}")
print(f"NumPy's exp({x_val}) = {numpy_exp_val}")
print(f"Difference: {abs(custom_exp_val - numpy_exp_val)}")


# Truncation Level Analysis


trunc_levels = range(1, 21)
differences = []

for N in trunc_levels:
    custom_result = exp(x_val, N=N)
    numpy_result = np.exp(x_val)
    difference = abs(custom_result - numpy_result)
    differences.append(difference)

plt.plot(trunc_levels, differences)
plt.xlabel('Truncation Level (N)')
plt.ylabel('Absolute Difference from NumPy Result')
plt.title('Truncation Level vs Error for exp()')
plt.show()
```

This difference is extremely small, on the order of machine precision for double-precision floating-point numbers.

This indicates that the custom implementation is highly accurate for this input value when compared to the trusted NumPy's implementation. The discrepancy is likely due to minute rounding errors inherent in floating-point arithmetic and is typically considered negligible for most practical applications.

## Sin Function

```python
# Testing the sin() function
x_val = 200*np.pi/4  # 45 degrees in radians
custom_sin = sin(x_val,20)
numpy_sin = np.sin(x_val)

print(f"Custom sin({x_val}) = {custom_sin}")
print(f"NumPy's sin({x_val}) = {numpy_sin}")
print(f"Difference: {abs(custom_sin - numpy_sin)}")

# This process should be repeated for different values of x_val to test accuracy across a range of inputs.


# Truncation Level Analysis


# Analyze the effect of the truncation level on the sin() function.

trunc_levels = range(1, 21)  # for example, testing 1 to 20
x_val = 20*np.pi/4  # or any suitable test value

differences = []

for level in trunc_levels:
    custom_result = sin(x_val, N=level)
    numpy_result = np.sin(x_val)
    difference = abs(custom_result - numpy_result)
    differences.append(difference)

plt.plot(trunc_levels, differences)
plt.xlabel('Truncation Level')
plt.ylabel('Absolute Difference from NumPy Result')
plt.title('Truncation Level vs Error for sin() function')
plt.show()
```

This is again a very small value, close to machine precision for double-precision floating-point numbers.

Considering the input value is approximately 5π, the expected sine value is close to zero. Both custom and NumPy implementations return values very near to zero, but there's a small discrepancy due to potential rounding errors, truncation errors, or the specifics of the algorithms employed.
